# The fractional Fourier transform

The fractional Fourier transform (FrFT) is a time-frequency analysis technique that extends the traditional Fourier transform (FT) to the entire time-frequency domain by using fractional powers of the FT operator $\mathcal{F}$. The FrFT generalizes the FT with an extra parameter $a$ that can be related to a signal rotation in the time-frequency plane. Under such scheme, the temporal signal is decomposed using a chirp function basis. Thus, the FrFT provides a unified time-frequency representation of the signal with higher time-frequency resolution than other techniques.

The FrFT has multiple applications in quantum mechanics, optics and signal processing. It has been widely used in the optics field, due to the possibility of its practical implementation. Developments in speech, music, biomedical signals, and mechanical vibrations are among the signal processing solutions that this tool can further potentiate.

# The Fractional Fourier Transform

In time-frequency analysis, a signal is represented in a plane where the time and frequency domains are orthogonal axes as depicted by the dashed lines in the figure below.

For a signal $x(t)$ along the time axis, its traditional FT $X(\omega)$ corresponds to the frequency axis, meaning a $90^{\circ}$ rotation in the time-frequency plane. The FrFT of real order $a$, designated by the $\mathcal{F}^a$ operator, enables intermediate representations between time and frequency domains. Therefore, the FrFT defines a signal along the $u_a$ axis, also known as the fractional axis, that yields a unified time-frequency characterization. The $u_a$ is a rotated axis within the time-frequency plane by an angle $\phi$ as illustrated by the solid lines

<img src="../../reports/figures/frft_axes_rotation.png" width="400">

The $a$-th-order FrFT of a signal $x(u_0)$, where $u_0$ represents the initial domain, i.e., time, space, frequency or even an intermediate domain, is defined as follows:

$$
\mathcal{F}^a\{x(u_0)\} = X(u_a) = \int\limits_{-\infty}^\infty K_a(u_0,u_a) x(u_0) \mathrm{d}u_0.
$$

The kernel $K_a(u_0,u_a)$ is given by the following expression:

$$
K_a(u_0,u_a) = 
\left\{
\begin{aligned}
& A_\phi e^{[j\pi(u_a^2 \cot{\phi} - 2 u_a u_0 \csc{\phi} + u_0^2 \cot{\phi})]},\\
& \delta(u_a-u_0) \quad \text{if} \quad a \to 4n,\\
& \delta(u_a+u_0) \quad \text{if} \quad a \to 4n \pm 2,
\end{aligned}
\right.
$$

where $j^2=-1$, $n \in \mathbb{Z}$ and:

$$
A_\phi = \sqrt{1-j\cot{\phi}},
$$

$$
\phi = \frac{a\pi}{2}.
$$

The square root in $A_\phi$ is calculated such that the argument of the result lies in the interval $(-\pi/2,\pi/2]$. The last expression establishes the relation between the rotation angle of the time-frequency plane $\phi$ and the fractional order $a$.

The relation of fractional domain $(u_0, u_a)$ with the time-frequency plane can be expressed as

$$
\begin{aligned}
u_a &= u_0 \cos{\phi} + u_1 \sin{\phi} \\
u_{a+1} &= -u_0 \sin{\phi} + u_1 \cos{\phi}
\end{aligned}
$$

which can be written as

$$
\begin{pmatrix}
u_a \\ u_{a+1}
\end{pmatrix} = 
\begin{pmatrix}
\cos{\phi} & \sin{\phi} \\
-\sin{\phi} & \cos{\phi}
\end{pmatrix}
\begin{pmatrix}
u_0 \\ u_1
\end{pmatrix}
$$

## Properties of the FrFT

- **Linearity**:

    $$
    \mathcal {F}^a \left[\sum \nolimits_k b_k f_k(u)\right] = \sum \nolimits_k b_k \mathcal{F}^a \left[f_k(u)\right]
    $$

- **Integer Orders**: When $a$ is an integer power, the $\mathcal{F}^a$ is equivalent to the ordinary fourier transform operator. Repeated aplication of the FT operator produce,

    $$
    \begin{aligned}
    \mathcal{F}^0\{x(u_0)\} &= x(u_0) \longrightarrow \mathcal{F}^0 = \mathcal{I}, \\
    \mathcal{F}^1\{x(u_0)\} &= X(u_1) \longrightarrow \mathcal{F}^1 = \mathcal{F}, \\
    \mathcal{F}^2\{x(u_0)\} &= x(-u_0) \longrightarrow \mathcal{F}^2 = \mathcal{P},\\
    \mathcal{F}^3\{x(u_0)\} &= X(-u_1) \longrightarrow \mathcal{F}^3 = \mathcal{P}\mathcal{F},\\
    \mathcal{F}^4\{x(u_0)\} &= x(u_0) \longrightarrow \mathcal{F}^4 = \mathcal{I}.
    \end{aligned}
    $$

    Then $\mathcal{F}^j = \mathcal{F}^{j\pmod{4}}$

- **Inverse**

    $$
    (\mathcal{F}^a)^{-1} = \mathcal{F}^{-a}
    $$

- **Unitarity**: The conjugate transpose of $\mathcal{F}^a$ is equal to its inverse. That is

    $$
    (\mathcal{F}^a)^{-1} = (\mathcal{F}^a)^H = \mathcal{F}^{-a}
    $$

- **Index additivity**

    $$
    \mathcal{F}^{a_1} \mathcal{F}^{a_2} = \mathcal{F}^{a_1 + a_2}
    $$

- **Commutativity**

    $$
    \mathcal{F}^{a_1} \mathcal{F}^{a_2} = \mathcal{F}^{a_2} \mathcal{F}^{a_1}
    $$

- **Associativity**

    $$
    \mathcal{F}^{a_1} (\mathcal{F}^{a_2} \mathcal{F}^{a_3}) = (\mathcal{F}^{a_1} \mathcal{F}^{a_2})\mathcal{F}^{a_3} 
    $$

- **Eigenfunctions**

    $$
    \mathcal{F}^a[\psi_n(u_0)] = e^{-j a n \pi/2}\psi_n(u_0)
    $$

- **Parseval**

    $$
    \langle f(u_0), g(u_0) \rangle = \langle f(u_a), g(u_a) \rangle.
    $$

    That is

    $$
    \int f^*(u_0)g(u_0)du_0 = \int f^*(u_a)g(u_a)du_a
    $$

    as a special case 

    $$
    \langle f(u_0), f(u_0) \rangle = \langle f(u_a), f(u_a) \rangle.
    $$

    which implies energy or norm conservation. That is 

    $$
    \|f(u_0)\| = \|f(u_a)\|
    $$

    and

    $$
    En[f(u_0)] = En[f(u_a)]
    $$

## Basis functions of the FrFT

The basis functions of the FrFT are linear frequenci modulated functions also known as *linear chirp*. These functions can be understood as signals characterized by their instantaneous frequency which varies linearly with time as describes the following equation:

$$
f(t) = ct + f_0
$$

where $f_0$ is the intantaneous frequency at time $t = 0$ and $c$ is the gradient or rate of frequency change for a given time interval. The gradient is defined as:

$$
c = \frac{\Delta f}{\Delta t} = \frac{f_1 - f_0}{T}
$$

where T is the period that the instantaneous frequncy takes to sweep from $f_0$ to $f_1$.

The corresponding time-domain function for the phase of a linear chirp is then,

$$
\varphi(t) = \varphi_0 + 2\pi \int\limits_0^t (c\tau + f_0) \mathrm{d}\tau.
$$

Solving the integral we obtain,

$$
\varphi(t) = \varphi_0 + 2\pi \left(\frac{ct^2}{2} + f_0t \right).
$$

Thus, the analytical expression for a linear frequency modulated signal is

$$
x(t) = A e^{j(\varphi(t))} = A e^{j\left[\varphi_0 + 2\pi \left(\frac{ct^2}{2} + f_0t \right)\right]}
$$

A plot of a linear chirp is depicted as follows 

We can analyse the structure of the FrFT kernel to stablish its relation with the linear chirp equation. Consider the expression in the exponent of the FrFT kernel

$$
\varphi_1(u_0, u_a) = \pi(u_a^2 \cot{\phi} - 2 u_a u_0 \csc{\phi} + u_0^2 \cot{\phi})
$$

We can write this expression in terms of one variable and substitute $u_0 = t$

$$
\begin{aligned}
\varphi_1(t) &= \pi(\cot{\phi} - 2 t \csc{\phi} + t^2 \cot{\phi}) \\
&= \pi \cot{\phi} - 2 \pi \csc{\phi} t  + \pi \cot{\phi}t^2 \\
\end{aligned}.
$$

On the other hand, the phase of the linear chirp can be writen as

$$
\varphi_2(t) = \varphi_0 + 2 \pi f_0 t + \pi ct^2.
$$

Both expressions $\varphi_1(t)$ and $\varphi_2(t)$ are quadratic polynomials on the varible $t$. If we relate their coeficients we find that

$$
\varphi_0 = \pi \cot{\phi}, \quad f_0 = - \csc{\phi}, \quad c = \cot{\phi}.
$$

From the las exprision we have that

$$
\frac{f_1 - f_0}{T} = \cot{\phi},
$$

as $\cot{\phi} = 1/\tan{\phi}$ we can write

$$
\frac{T}{f_1 - f_0} = \tan{\phi},
$$

and solving for $\phi$ we obtain

$$
\phi = \tan^{-1}{\left(\frac{T}{f_1 - f_0}\right)}.
$$

From the definition of the FrFT we have that

$$
\frac{a\pi}{2} = \tan^{-1}{\left(\frac{T}{f_1 - f_0}\right)}.
$$

Solvin for $a$ we get

$$
\begin{aligned}
a &= \frac{2}{\pi} \tan^{-1}{\left(\frac{T}{f_1 - f_0}\right)} \\
&= \frac{2}{\pi}\phi
\end{aligned}
$$

This expression proves the relation of the order of the FrFT with the rate of frequency change on a period $T$. Notice that, when $\phi \to 0$, $c \to \infty$, and when $\phi \to \pi$, $c \to -\infty$. We can illustrate this situation considering the rate of frequency change $c$. If $c \to \infty$ the change in frequency $\Delta f \to \infty$. Thus the chip signal takes the form of a Dirac delta function. Similarly, if $c \to -\infty$ the change in frequency $\Delta f \to -\infty$. Thus the chip signal takes the form of a Dirac delta function of inverted sequence. This can be evidence in the definition of the kernel of the FrFT.

# The FrFT of a linear chirp

Consider the following chip signal with $f_0 = 0$, $f_1 = 50$, $T = 1$, $\phi_0 = 0$ and a duration of one second.

```python
%load_ext autoreload
%autoreload 2
```

```python
# Requirements
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from dsp.signals.fm import lfmsig
from dsp.transforms.frft import frft

sns.set_style("whitegrid")
```

```python
fs = 1000
dur = 1
t = np.linspace(0, dur, fs)
f0 = 0
f1 = 50
T = 1

y = lfmsig(
    t,
    f0=f0,
    f1=f1,
    sweep_time=T,
    amp=1
)

n = len(y)

fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(t, np.real(y), label='Real part')
ax.plot(t, np.imag(y), label='Imag. part')
ax.set_ylim((-1.5, 1.5))
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.legend()
plt.show()
```

The following plot show the energy concentration of the chirp signal in four different fractional domains.

```python
# Compute FrFT for several orders and visualize energy concentration
# Frequency axis for plotting (not strictly frequency for FrFT, but useful for visualization)
f = np.linspace(0, fs, n)

# Fractional orders to analyze
a_orders = np.array([0.5, 0.8, 1, 1.03])
n_orders = len(a_orders)

# Compute FrFT for each order
Fa = np.zeros((n, n_orders), dtype=np.complex128)
for i, a_val in enumerate(a_orders):
    Fa[:, i] = frft(y, a_val)

# Plot squared modulus (energy) for each fractional domain
fig, axes = plt.subplots(2, 2, figsize=(12, 6))
axes = axes.flatten()
for i, a_val in enumerate(a_orders):
    axes[i].plot(f, np.abs(Fa[:, i])**2, label=f'a = {a_val}')
    axes[i].set_title(f'FrFT Energy, a = {a_val}')
    axes[i].set_ylabel('Squared modulus')
    axes[i].set_xlabel('Fractional frequency (Hz)')
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
